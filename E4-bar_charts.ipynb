```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [i * 100 for i in [13, 14, 16, 19, 21, 23]]
y = [i * 1000 for i in [88, 72, 94, 86, 112, 98]]
```

```python
plt.plot(x, y, 'o')
plt.xlabel('square feet')
plt.ylabel('house price')
plt.show()
```

```python
y_bar = [np.mean((y[i], y[i+1])) for i in (0, 2, 4)]
x_bar = [x[i] for i in (0, 2, 4)]
print(y_bar)
print(x_bar)
```

```python
plt.bar(x_bar, y_bar, width=300)
plt.show()
```
