```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1700, 2100, 1900, 1300, 1600, 2200]
y1 = [51000, 63000, 57000, 39000, 48000, 66000]
y2 = [53000, 65000, 59000, 41000, 50000, 68000]
y3 = [53000, 44000, 59000, 82000, 50000, 68000]
```

```python
plt.plot(x, y1, 'bo', x, y2, 'ro', x, y3, 'go')
plt.ylabel('house price')
plt.xlabel('square feet')
plt.show()
```

```python
slope, intercept = np.polyfit(x, y2, 1)
print(slope, intercept)
```
