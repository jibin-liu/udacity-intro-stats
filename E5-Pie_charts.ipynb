```python
import matplotlib.pyplot as plt
```

```python
labels = ['Party A', 'Party B']
sizes = [50, 50]
plt.pie(sizes, labels=labels, startangle=90, shadow=True, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
election = [724000, 181000]
sizes = [i / sum(election) for i in election]
party = ['Party A', 'Party B']
plt.pie(sizes, labels=party, colors=['red', 'blue'], autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```

```python
students = [12000, 96000, 36000]
ages = ['13-19', '20-32', '33-999']
colors = ['blue', 'red', 'black']
plt.pie(students, labels=ages, colors=colors, autopct="%1.1f%%", startangle=90)
plt.axis('equal')
plt.show()
```

```python
party = [175000, 50000, 25000, 50000]
colors = ['black', 'blue', 'red', 'green']
labels = list('ABCD')
res = plt.pie(party, colors=colors, labels=labels, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
pct = [float(text.get_text()[:-1]) for text in res[-1]]
total = 240000
numbers = [total * i / 100 for i in pct]
print(numbers)
```
